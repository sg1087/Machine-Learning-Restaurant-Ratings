```python
import pandas as pd 
```

```python
df1 = pd.read_csv('Final_Rest_Data.csv')
df1
```

```python
df1= df1.rename(index=str, columns={'aggregate_rating': 'User Rating','votes': 'Number of Reviews','establishment':'Establishment','Price':'Price Range'})
df1
```

```python
final_data = df1.drop(['Restaurant_x'], axis=1)
final_data
```

```python
print(final_data['Number of Reviews'].mean())
print(final_data['Number of Reviews'].std())
```

```python
#for number of reviews greater than mean
final_data['Reviews'] = final_data['Number of Reviews'].apply(lambda x: 'low' if x < 22 else 'high')
```

```python
final_data
```

```python
final_data['Rating'] = final_data['User Rating'].astype(int)
final_data['Rating'] = final_data['Rating'].astype(str)
final_data
```

```python
final_data_clean = final_data.drop(['User Rating','Number of Reviews'], axis=1)
final_data_clean
```

```python
import arff
arff.dump('final-restaurants-cleaned.arff'
      , final_data_clean.values
      , relation='final-restaurants-cleaned'
      , names=final_data_clean.columns)
```
