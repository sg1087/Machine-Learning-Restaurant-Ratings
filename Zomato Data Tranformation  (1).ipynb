```python
import pandas as pd 
```

```python
zomato_data = pd.read_csv('./Downloads/zomato_output.csv')
zomato_data
```

```python
zomato_data = zomato_data.rename(index=str, columns={'aggregate_rating': 'User Rating','votes': 'Number of Reviews','establishment':'Establishment'})
zomato_data
```

```python
zomato_data_clean = zomato_data[['Borough','Grade','Number of Reviews','Establishment','User Rating']]
zomato_data_clean
```

```python
zomato_data_clean['Number of Reviews'] = zomato_data_clean['Number of Reviews'].astype(int)
zomato_data_clean['User Rating'] = zomato_data_clean['User Rating'].astype(float)
```

```python
zomato_data_clean['Number of Reviews'].mean()
```

```python
zomato_data_clean['Reviews'] = zomato_data_clean['Number of Reviews'].apply(lambda x: 'low' if x < 15 else 'high')
```

```python
zomato_data_clean
```

```python
zomato_data_clean['Rating'] = zomato_data_clean['User Rating'].round()
zomato_data_clean
```

```python
zomato_data_clean['Rating'] = zomato_data_clean['Rating'].astype(str)
```

```python
zomato_data_transform= zomato_data_clean[['Borough','Grade','Reviews','Establishment','Rating']]
zomato_data_transform
```
